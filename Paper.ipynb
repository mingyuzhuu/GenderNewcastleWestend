```python
df = pd.read_csv('CFeature.csv')
```

```python
# Remove the unnecessary columns
df = df.drop(columns=['BName','SName','Loc','BusName','SchName','RName','PubName','id','GName'])
```

```python
# Move 'count' to the front column 
columns = ['count'] + [col for col in df.columns if col != 'count']
df = df[columns]
df.head()
```

```python
# Identify the columns to move (10th to 15th)
columns_to_move = columns[9:16]
# Remove these columns from the original list
remaining_columns = [col for i, col in enumerate(columns) if i not in range(9, 16)]
# Identify the insertion point (before the 8th column)
insertion_point = 7
# Reconstruct the column order
new_column_order = remaining_columns[:insertion_point] + columns_to_move + remaining_columns[insertion_point:]
# Reorder the DataFrame columns
df = df[new_column_order]
df.head()
```

```python
# exlude the outlier
df = df[df['count'] <= 100]
```

## Percentage of POI within 200m of offence location

```python
data = {
    'Category': ['Barber(<200m)', 'Grocery(<200m)'
#                  ,'Bus(<200m)','School(<200m)'
                 ,'Restaurant(<200m)','Pub(<200m)','GreenSpace(<200m)' ],
    'Percentage': [(df[df['BDist'] < 200])['count'].sum()/df['count'].sum()
                   , (df[df['SDist'] < 200])['count'].sum()/df['count'].sum()
#                    , (df[df['BusDist'] < 200])['count'].sum()/df['count'].sum()
#                    , (df[df['SchDist'] < 200])['count'].sum()/df['count'].sum()
                   , (df[df['RDist'] < 200])['count'].sum()/df['count'].sum()
                   , (df[df['PubDist'] < 200])['count'].sum()/df['count'].sum()
                   , (df[df['PubDist'] < 200])['count'].sum()/df['count'].sum()]
}

df_per = pd.DataFrame(data)

# Define colors based on the condition
colors = ['red' if value > 0.49 else 'grey' for value in df_per['Percentage']]

# Creating the horizontal bar plot
plt.figure(figsize=(10, 3))
plt.barh(df_per['Category'], df_per['Percentage'],height=0.6, alpha=0.7, color=colors)

# Adding title and labels
plt.title('')
plt.xlabel('', fontsize=15)
plt.ylabel('', fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)

# Formatting the x-axis to show percentages
plt.gca().xaxis.set_major_formatter(mtick.PercentFormatter(1))

# Adding text annotations to show the percentages at the end of each bar
for index, value in enumerate(df_per['Percentage']):
    plt.text(value, index, f'{int(value * 100)}%', va='center'
             , ha='left',bbox=dict(facecolor='none', edgecolor='none'), fontsize=14, fontweight='bold')  # 'va' centers the text vertically

# Remove the black edge (spines) around the figure
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Save the plot as a high-resolution PNG file
# plt.savefig('percentagePOI.png', dpi=300, bbox_inches='tight')
      
# Display the plot
plt.show()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import numpy as np

# Sample data
data = {
    'Y': [1, 2, 3, 4, 5],
    'X1': [5, 4, 3, 2, 1],
    'X2': [2, 3, 4, 5, 6],
    'X3': [5, 3, 2, 4, 1]
}
df = pd.DataFrame(data)

# Define the dependent and independent variables
X = df[['X1', 'X2', 'X3']]
y = df['Y']

# Create and fit the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# View the coefficients and intercept
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)

# Make predictions (optional)
y_pred = model.predict(X)
print('Predictions:', y_pred)
```
